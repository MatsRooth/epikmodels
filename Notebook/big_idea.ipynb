````python
k = 1
for dname in EventNameU:
    dname = dname + "u"
    d = eval(dname)
    for ename in EventNameU:
        ename = ename + "u"
        e = eval(ename)
        de = Cn(d,e)
        print(f'{k}. Amy\'s unnormalized weighted alternatives to {dname} {ename}')
        image = amyWImage(de)
        image.push_weights_to_end()
        display(image.view())
        # See if we can sum to the input
        print('  Attempted sum:')
        de.compose(amyU)
        de.compose(image)
        de.input_project()
        de.push_weights_to_end()
        display(de.view())
        de.determinize()
        print('  Determinized attempted sum:')
        display(de.view())
        k = k + 1
````

````python
print('Weighted alternatives mapped back to base.')
b1 = w1.copy()
b1.compose(amyU)
b1.compose(w1_image)
b1.input_project()
b1.push_weights_to_end()
display(b1.view())
#print(' ... and summed.')
#ww.convert(hfst.ImplementationType.LOG_OPENFST_TYPE)
#ww.determinize()
#display(ww.view())
````
